```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
sns.set_theme(style="ticks", color_codes=True)
from tqdm.notebook import tqdm, trange
import time
```

```python
start_time = time.time()

main_01 = pd.read_csv('RAW/data60.csv')
main_02 = pd.read_csv('RAW/data61.csv')
main_03 = pd.read_csv('RAW/data62.csv')
main_04 = pd.read_csv('RAW/data63.csv')
main_05 = pd.read_csv('RAW/data64.csv')
main_06 = pd.read_csv('RAW/data65.csv')
main_07 = pd.read_csv('RAW/data66.csv')
main_08 = pd.read_csv('RAW/data67.csv')
main_09 = pd.read_csv('RAW/data68.csv')
main_10 = pd.read_csv('RAW/data69.csv')
main_11 = pd.read_csv('RAW/data70.csv')
main_12 = pd.read_csv('RAW/data71.csv')

end_time = time.time()
total_time = end_time - start_time

print("运行时间：", total_time, "秒")
```

```python
all_data = [main_01,main_02,main_03,main_04,main_05,main_06,main_07,main_08,main_09,main_10,main_11,main_12]
```

```python
cleaned_all_data = []

for data in all_data:
    cleaned_all_data.append(data.groupby(['日期','時段','出站']).sum().reset_index())

full_year_data = pd.concat(cleaned_all_data)
full_year_data['日期'] = pd.to_datetime(full_year_data['日期'])
```

```python
full_year_data.rename(columns = {'日期': 'Date', '時段': 'Hour', '出站':'Station','人次':'CrowdFlow'}, inplace = True)
full_year_data = full_year_data.reset_index().drop(['index'],axis=1)
full_year_data
full_year_data.to_csv('2022.csv',  index=False, encoding='utf-8')
```

```python
ppl_every_day = full_year_data.groupby(['Date']).sum().reset_index().drop(['Hour'],axis=1)
ppl_every_day
```

```python
ppl_every_day = full_year_data.groupby(['Date']).sum().reset_index()

fig = plt.figure(figsize=(20,20))
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(ppl_every_day['Date'][:365], ppl_every_day['CrowdFlow'][:365])
plt.show()
```

```python

ppl_per_station = full_year_data.groupby(['Station']).sum().reset_index()
flow = ppl_per_station['CrowdFlow'].tolist()

%matplotlib inline
plt.style.use('fivethirtyeight')
f, ax = plt.subplots(figsize=(20,20)) # set the size that you'd like (width, height)
x_values = list(range(len(flow)))
plt.bar(x_values, flow, orientation = 'vertical')
station_list = ppl_per_station['Station'].tolist()
plt.xticks(x_values, station_list, rotation='vertical')
plt.ylabel('Crowd flow'); plt.xlabel('Stations'); plt.title('2 Years crowd flow per station');
```

```python
station_ranking = ppl_per_station.sort_values(by=['CrowdFlow'],ascending=False)
station_ranking = station_ranking.reset_index().drop(['index'],axis=1)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(station_ranking)
```
